# Check for null values in the numerical columns 

```python
numericals = data.select_dtypes(include = np.number)
numericals.isna().sum()
```

```python
nulls_num_df = pd.DataFrame(numericals.isna().sum()/len(numericals)).reset_index()
nulls_num_df.columns = ['name', 'nulls_percentage']
nulls_num_df[nulls_num_df['nulls_percentage']!= 0]
```

```python
above_threshold = nulls_num_df[nulls_num_df['nulls_percentage']>0.25]                    #look at columns with nulls above 25%
above_threshold['name']
```

```python
above_threshold.shape
```

```python
drop_list = list(above_threshold['name'])           
print(drop_list)
```

```python
# drop_list_num = []
# for col_name in drop_list:
#     if "ADATE" in col_name or 'RDATE' in col_name or 'RAMNT' in col_name:
#         drop_list_num.append(col_name)   
```

```python
numericals = numericals.drop(columns=drop_list)
numericals.head()
```

```python
numericals['AGE'] = numericals["AGE"].fillna(np.mean(numericals['AGE']))
sns.distplot(numericals['AGE'])
plt.show()
```

```python
sns.distplot(numericals[numericals['INCOME'].isna()==False]['INCOME']) 
plt.show()
```

```python
print(numericals['INCOME'].value_counts())      #it's categorical!
numericals['INCOME'] = numericals['INCOME'].astype('object')
```

```python
numericals['INCOME'] = numericals['INCOME'].fillna('5.0') # Replacing the null values with the most represented categoty
```

```python
sns.distplot(numericals[numericals['CLUSTER2'].isna()==False]['CLUSTER2']) 
plt.show()
```

```python
numericals['CLUSTER2'] = numericals['CLUSTER2'].fillna(np.ceil(np.mean(numericals['CLUSTER2'])))
```

# Use appropriate methods to clean the columns GEOCODE2, WEALTH1, ADI, DMA,and MSA.

```python
#GEOCODE2
data['GEOCODE2'].value_counts()
def clean(x):
    if x == ' ':
        return 'A'
    else:
        return x
data['GEOCODE2'] = data['GEOCODE2'].apply(clean)
data['GEOCODE2'] = data['GEOCODE2'].fillna('A')    #fill with 'A'
data['GEOCODE2'].value_counts(dropna=False)
```
