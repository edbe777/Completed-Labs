## Remove those columns from the dataframe

```python
def drop_columns(df, drop_list):
    for i in drop_list:
        df = df.drop(i, axis = 1)
    return df
```

```python
drop_columns = drop_columns(data, ['rdate_5', 'ramnt_5', 'rdate_3', 'ramnt_3', 'rdate_4', 'ramnt_4',
       'ramnt_6', 'rdate_6', 'ramnt_15', 'rdate_15', 'rdate_23',
       'ramnt_23', 'rdate_20', 'ramnt_20', 'ramnt_7', 'rdate_7',
       'ramnt_17', 'rdate_17', 'rdate_21', 'ramnt_21', 'ramnt_10',
       'rdate_10', 'ramnt_13', 'rdate_13', 'numchld'])
```

## Reduce the number of categories in the column GENDER. The column should only have either "M" for males, "F" for females, and "other" for all the rest

#Note that there are a few null values in the column. We will first replace those null values using the code below:

```python
print(data['gender'].value_counts())
data['gender'] = data['gender'].fillna('F')
```

```python
#check if filled out correctly
data['gender'].value_counts() 
```

```python
#check for space
data['gender'].unique()
```

```python
#replace space with 'F'
data['gender'] = data['gender'].replace(' ', 'F')
data['gender'].value_counts()
```

```python
#use lambda in gender to separate into 3 categories: M, F, other
data['gender'] = data['gender'].apply(lambda x: x if x in ['M', 'F'] else 'other')
data['gender'].value_counts()
```
