```python
import pandas as pd

file1 = pd.read_csv('./file1.csv')
display(file1.head())
file2 = pd.read_csv('./file2.csv')
display(file2.head())
file3 = pd.read_csv('./file3.csv')
display(file3.head())
```

```python
display(file1.shape)
display(file2.shape)
display(file3.shape)
```

```python
cols1 = []
for colname in file1.columns:
    cols1.append(colname.lower())
file1.columns = cols1
cols1
```

```python
cols2 = []
for colname in file2.columns:
    cols2.append(colname.lower())
file2.columns = cols2
cols2
```

```python
cols3 = []
for colname in file3.columns:
    cols3.append(colname.lower())
file3.columns = cols3
cols3
```

```python
display(file1.head())
display(file2.head())
display(file3.head())
```

```python
file2 = file2[['customer',
 'st',
 'gender',
 'education',
 'customer lifetime value',
 'income',
 'monthly premium auto',
 'number of open complaints',
 'policy type',
 'vehicle class',
 'total claim amount']]
file2.head()
```

```python
file3 = file3.rename(columns={'state':'st'})
```

```python
file3 = file3[['customer',
 'st',
 'gender',
 'education',
 'customer lifetime value',
 'income',
 'monthly premium auto',
 'number of open complaints',
 'policy type',
 'vehicle class',
 'total claim amount']]
file3.head()
```

```python
data = pd.concat([file1,file2,file3])
data
```

```python
data.dtypes
```

```python

data_copy = data.drop(['education','number of open complaints'], axis=1) 
data_copy.head()
```

```python
data_copy['customer lifetime value'] = data_copy['customer lifetime value'].astype(str)
data_copy['customer lifetime value'] = list(map(lambda x: x.replace('%',''), data_copy['customer lifetime value']))
data_copy
```

```python
data_copy['customer lifetime value'] = data_copy['customer lifetime value'].astype('float')/100
data_copy
```

```python
display(data_copy.shape)
data_copy = data_copy.drop_duplicates()
display(data_copy.shape)
```
