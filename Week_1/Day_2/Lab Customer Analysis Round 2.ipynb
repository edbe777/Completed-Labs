```python
import pandas as pd
import numpy as np

file = pd.read_csv('./marketing_customer_analysis.csv')
display(file.head())
display(file.shape)
```

```python
cols = []
for colname in file.columns:
    cols.append(colname.lower())
file.columns = cols
cols
```

```python
display(file.head())
```

```python
#file = file.rename(columns={'unnamed: 0':'unnamed'})
file = file.drop(['unnamed: 0'], axis=1)
file.head()
```

```python
file = file.replace(np.nan,0)
file.head()
```

```python
file._get_numeric_data()
```

```python
file.select_dtypes(object)
```

```python
file['date time']= pd.to_datetime(file['effective to date'])
file['months'] = file['date time'].dt.month_name()
file.head()
```

```python
file = file[(file['months'] == 'January') | (file['months'] == 'February') | (file['months'] == 'March')]
display(file.head())
display(file.shape)
```
